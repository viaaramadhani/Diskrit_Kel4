```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
```

```python
x_fabric = np.arange(0, 11, 1)
x_wrinkle = np.arange(0, 11, 1)
x_temp = np.arange(0, 301, 1)
```

```python
fabric_delicate = fuzz.trimf(x_fabric, [0, 0, 5])
fabric_normal = fuzz.trimf(x_fabric, [0, 5, 10])
fabric_heavy = fuzz.trimf(x_fabric, [5, 10, 10])

wrinkle_low = fuzz.trimf(x_wrinkle, [0, 0, 5])
wrinkle_medium = fuzz.trimf(x_wrinkle, [0, 5, 10])
wrinkle_high = fuzz.trimf(x_wrinkle, [5, 10, 10])

temp_low = fuzz.trimf(x_temp, [0, 0, 150])
temp_medium = fuzz.trimf(x_temp, [0, 150, 300])
temp_high = fuzz.trimf(x_temp, [150, 300, 300])
```

```python
fig, (ax0, ax1, ax2) = plt.subplots(nrows=3, figsize=(8, 9))

ax0.plot(x_fabric, fabric_delicate, 'b', linewidth=1.5, label='Delicate')
ax0.plot(x_fabric, fabric_normal, 'g', linewidth=1.5, label='Normal')
ax0.plot(x_fabric, fabric_heavy, 'r', linewidth=1.5, label='Heavy')
ax0.set_title('Fabric Type')
ax0.legend()

ax1.plot(x_wrinkle, wrinkle_low, 'b', linewidth=1.5, label='Low')
ax1.plot(x_wrinkle, wrinkle_medium, 'g', linewidth=1.5, label='Medium')
ax1.plot(x_wrinkle, wrinkle_high, 'r', linewidth=1.5, label='High')
ax1.set_title('Wrinkle Level')
ax1.legend()

ax2.plot(x_temp, temp_low, 'b', linewidth=1.5, label='Low')
ax2.plot(x_temp, temp_medium, 'g', linewidth=1.5, label='Medium')
ax2.plot(x_temp, temp_high, 'r', linewidth=1.5, label='High')
ax2.set_title('Iron Temperature')
ax2.legend()
```

```python
for ax in (ax0, ax1, ax2):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()
```

```python
plt.tight_layout()
plt.show()
```
